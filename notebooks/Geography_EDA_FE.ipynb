# Feature Engineering


### Tách City thành hai cột: City_Name và State_Code

```python
df_geo[['City_Name', 'State_Code']] = df_geo['City'].str.split(',', expand=True)
df_geo['City_Name'] = df_geo['City_Name'].str.strip()
df_geo['State_Code'] = df_geo['State_Code'].str.strip()
```

### Mã hóa Region và District bằng Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

label_encoder_region = LabelEncoder()
label_encoder_district = LabelEncoder()

df_geo['Region_Encoded'] = label_encoder_region.fit_transform(df_geo['Region'])
df_geo['District_Encoded'] = label_encoder_district.fit_transform(df_geo['District'])
```

### Tạo Zip prefix (3 chữ số đầu tiên của Zip)

```python
df_geo['Zip_Prefix'] = df_geo['Zip'].astype(str).str[:3]
```
